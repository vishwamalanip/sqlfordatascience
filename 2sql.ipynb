               (sid | bcode)

            FROM orders o

    Find the restaurant with max repeat customers
        --> SELECT r.r_name,COUNT(*) AS loyal_customers 
            FROM (SELECT r_id,user_id,COUNT(*) AS visits 
                  FROM orders 
                  GROUP BY r_id,user_id 
                  HAVING visits>1) t 
                  JOIN restaurants r
                  ON r.r_id=t.r_id
                  GROUP BY t.r_id
                  ORDER BY loyal_customers DESC LIMIT 1;
                  
    Month over month revenue growth of a restaurant/swiggy
        --> SELECT month,((revenue - prev)/prev)*100 FROM(
                WITH sales AS
                      (
                        SELECT MONTHNAME(date) AS 'month',SUM(amount) AS 'revenue'
                        FROM orders 
                        GROUP BY month
                        ORDER BY MONTH(date)
                      )

                SELECT month,revenue,LAG(revenue,1)
                OVER(ORDER BY revenue) AS prev 
                FROM sales
                ) t
